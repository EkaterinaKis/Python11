```python
#1. Определить корни

from sympy.plotting import plot
from sympy import Symbol, solveset, diff, solve, pprint, S, sin, cos
x = Symbol('x')
f = (sin(x)**2) - (cos(x)**2)
solveset(f,x)
```

```python
#2. Найти интервалы, на которых функция возрастает

diff_f = diff(f,x)
pprint(solveset(diff_f>0,x,domain=S.Reals), use_unicode=True)
```

```python
#3. Найти интервалы, на которых функция убывает

pprint(solveset(diff_f<0,x,domain=S.Reals), use_unicode=True)
```

```python
#4. Построить график

plot(f)
```

```python
#5. Вычислить вершину

print(diff_f) #производная
print(solve(diff_f,x)) #вершина
```

```python
#6. Определить промежутки, на котором f > 0

pprint(solveset(f>0,x,domain=S.Reals), use_unicode=True)
```

```python
#7. Определить промежутки, на котором f < 0

pprint(solveset(f<0,x,domain=S.Reals), use_unicode=True)
```
